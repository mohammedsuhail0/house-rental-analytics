# Part 2: Exploring Our Data (Drawing Pictures!) 🎨
In this notebook, we act like detectives! We will draw pictures (called graphs) to see patterns in our clean data.


```python
# Import pandas to read the table, and matplotlib/seaborn to draw pictures!
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read our Clean Data
Let's open the file we saved in the last notebook!


```python
df = pd.read_csv('Cleaned_House_Data.csv')
df.head()
```

## Graph 1: How big are the houses?
Let's make a chart showing how many small houses vs big houses there are.


```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Area_in_Gaj'], bins=30, color='blue')
plt.title('Sizes of Houses in Hyderabad')
plt.xlabel('Size of House (in Gaj)')
plt.ylabel('How many houses?')
plt.show()
```

## Graph 2: Does a bigger house cost more money?
Let's put dots on a graph. The wider the dot goes, the bigger the house. The higher the dot goes, the more expensive it is!


```python
plt.figure(figsize=(8, 5))
# We use a 'scatterplot' for dots!
sns.scatterplot(data=df, x='Area_in_Gaj', y='Price', color='orange', alpha=0.6)
plt.title('House Prices vs Size')
plt.xlabel('Size of House (in Gaj)')
plt.ylabel('Price (in Rupees)')
plt.show()

# Notice how the dots generally go UP as they go RIGHT? That means bigger = more expensive!
```
